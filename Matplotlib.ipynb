```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#usine linspace function for the equal values
x=np.linspace(0,20,100)
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
print(y)
print(z)
```

```python
#sine curve
plt.plot(x,y)
```

```python
#tangent curve
plt.plot(x,z)
plt.show()
```

```python
#adding labels to graph
plt.plot(x,y)
plt.xlabel("values")
plt.ylabel("angle")
plt.title("Sine Curve")
```

```python
#Cirle
a=np.linspace(0,10,100)
b=np.linspace(0,10,100)
c=a**2+b**2
plt.plot(c)
```

```python
plt.plot(c,'.')
```

```python
d=(23,20,27,29,30,33,35,40)
e=(3,4,5,6,7,8,9,10)
plt.plot(d,e)
```

```python
#plotting 2 graphs in 1 
plt.plot(x,np.tan(x),'y')
plt.plot(x,np.sin(x),'g')
plt.show()
```

```python
#plotting a bar graph
fig=plt.figure()
gr=fig.add_axes([0,0,1,1])
State=['Andhra','Telanagna','Tamil nadu','Karnataka']
people=[100000,234444,122234,800000]
gr.bar(State,people)
```

```python
#pie chart
fig1=plt.figure()
gr=fig1.add_axes([0,0,1,1])
State=[' Andhra ','Telanagna','Tamil nadu','Karnataka']
people=[100000,234444,122234,800000]
gr.pie(people, labels=State, autopct='%1.1f%%')
```

```python
#Scatter plot
fig3=plt.figure()
gr=plt.axes(projection='3d')
z=30*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
gr.scatter(x,y,c=z,cmap='Greens')
plt.show()
```
